## La simulaci√≥n de Monte Carlo

* Generar dos numeros aleatorios uniforme entre (0,1) , X e Y en total 1000 veces
* Calcularemos X*X + Y*Y
    * Si el valor es inferior a 1 -> estamos dentro del c√≠rculo
    * Si el valor es superior a 1 -> estamos fuera del c√≠rculo
* calculamos el n√∫mero total de veces que est√°n dentro del c√≠rculo y lo dividimos entre el n√∫mero total de intentos para obtener una aproximaci√≥n de la probabilidad de caer dentro del c√≠rculo
* Usamos dicha probabilidad para aproximar elv alor de œÄ (Pi)
* Repetimos el experimento un n√∫mero suficiente de veces (por ejemplo 100), apra obtener (1000) diferentes aproximaciones de œÄ (Pi)
* Calculamos el promedio de los 100 experimentos para dar un valor final de œÄ (Pi)

```python
def pi_montecarlo(n,n_exp):
    pi_avg = 0
    pi_value_list = []
    for i in range(n_exp):
        value = 0
        x =np.random.uniform(0,1,n).tolist()
        y =np.random.uniform(0,1,n).tolist()
        for j in range(n):
            z = np.sqrt(x[j]* x[j]+y[j]*y[j])
            if z<=1:
                value +=1
        float_value = float(value)
        pi_value = float_value * 4 / n
        pi_value_list.append(pi_value)
        pi_avg += pi_value

    pi = pi_avg/n_exp
    print(pi)
    fig = plt.plot(pi_value_list)
    return(pi,fig)
```

```python
pi_montecarlo(10000,200)
```
