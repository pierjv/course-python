## Datos distribuidos

```python
import pandas as pd
data = pd.read_csv("../datasets/distributed-data/001.csv")
```

```python
data.head(10)
```

```python
data.shape
```

* importar el primer fichero , para cargar los datos y crear el dataset original
* crear un bucle para ir recorriendo los ficheros
* importante tener una consistencia en el nombre de ficheros
* cada uno debe apendizarse el fichero sobre lo que ya se cargaron
* repetimos hasta que no queden ficheros

```python
filepath = "../datasets/distributed-data/"
data = pd.read_csv("../datasets/distributed-data/001.csv")
final_lenght = len(data)

for i in range(2,333):
    if i < 10:
        filename = "00"+str(i)
    if 10 <= i < 100:
        filename = "0"+str(i)
    if i >= 100:
        filename = str(i)
    file = filepath+filename+".csv"
    temp_data = pd.read_csv(file) 
    final_lenght += len(temp_data)
    
    data = pd.concat([data,temp_data],axis=0)
    
```
