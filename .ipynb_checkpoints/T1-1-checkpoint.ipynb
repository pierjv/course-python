```python
data4 = open(mainpath +"/" + "customer-churn-model/Customer Churn Model.txt","r")
```

```python
cols = data4.readline().strip().split(",")
n_cols = len(cols)
n_cols
```

```python
counter=0
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
main_dict
```

```python
for line in data4:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter +=1
    
print("el data set tiene %d files y %d columnas" %(counter,n_cols))
```

```python
df3 = pd.DataFrame(main_dict)
df3.head()
```

# Lectura y escritura de ficheros

```python
infile = mainpath +"/" + "customer-churn-model/Customer Churn Model.txt"
outfile = mainpath +"/" + "customer-churn-model/Tab Customer Churn Model.txt"
```

```python
with open(infile,"r") as infile:
    with open(outfile,"w") as outfile:
        for line in infile:
            fields= line.strip().split(",")
            outfile.write("\t".join(fields))
            outfile.write("\n")
```

```python
df4 = pd.read_csv(outfile,sep="\t")
df4.head()
```
